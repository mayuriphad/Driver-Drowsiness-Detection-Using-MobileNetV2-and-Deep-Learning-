```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

# Define the classes (categories) and the directory path
Classes = ['Closed_eyes', 'Open_eyes']  # Example class names
Datadirectory = r"C:\Users\Mayuri\Desktop\Driver Drowsiness (2)\Driver Drowsiness\Test_dataset"  

# Read and display an image in grayscale
img_array = cv2.imread(r"C:\Users\Mayuri\Desktop\Driver Drowsiness (2)\Driver Drowsiness\Test_dataset\Closed_eyes\s0001_00001_0_0_0_0_0_01.png", cv2.IMREAD_GRAYSCALE)
plt.imshow(img_array, cmap="gray")
plt.show()

# Check the shape of the image
print(img_array.shape)

# Resize the image
img_size = 224
backtorgb = cv2.cvtColor(img_array, cv2.COLOR_GRAY2RGB)
new_array = cv2.resize(backtorgb, (img_size, img_size))
plt.imshow(new_array)
plt.show()

# Define training data list
training_data = []

def create_training_Data():
    for category in Classes:
        path = os.path.join(Datadirectory, category)
        class_num = Classes.index(category)  # Label each class
        print(f"Processing category: {category} Path: {path}")
        
        for img in os.listdir(path):
            try:
                img_path = os.path.join(path, img)
                img_array = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
                
                if img_array is None:
                    print(f"Could not read image: {img_path}")
                    continue
                
                backtorgb = cv2.cvtColor(img_array, cv2.COLOR_GRAY2RGB)
                new_array = cv2.resize(backtorgb, (img_size, img_size))
                training_data.append([new_array, class_num])
                
                print(f"Appended {img} from category {category}")
            except Exception as e:
                print(f"Error processing {img}: {e}")
                pass

# Call the function to create training data
create_training_Data()
print(f"Total training samples: {len(training_data)}")

# Shuffle the data
import random
random.shuffle(training_data)

# Prepare features and labels
X = []
y = []
for features, label in training_data:
    X.append(features)
    y.append(label)

# Convert to numpy arrays
X = np.array(X).reshape(-1, img_size, img_size, 3)
X = X / 255.0  # Normalize
Y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print(f"Training data shape: {X_train.shape}, Testing data shape: {X_test.shape}")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# Load the MobileNetV2 model without the top layers
base_model = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')

# Freeze the base model layers to avoid modifying the pretrained weights
base_model.trainable = False

# Add custom layers on top of the base model
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(1, activation='sigmoid')  # Binary classification (open/closed eyes)
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Model summary
model.summary()
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split

# Load the MobileNetV2 model without the top layers
base_model = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')

# Unfreeze some layers of the base model for fine-tuning
base_model.trainable = True
for layer in base_model.layers[:-50]:  # Adjust this value based on experimentation
    layer.trainable = False

# Add custom layers on top of the base model with dropout and L2 regularization
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dropout(0.5),  # Dropout with a rate of 50%
    layers.Dense(1, activation='sigmoid', kernel_regularizer=regularizers.l2(0.01))  # L2 regularization
])

# Compile the model with a lower learning rate
model.compile(optimizer=tf.keras.optimizers.Adam(1e-5), loss='binary_crossentropy', metrics=['accuracy'])

# Early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Split the dataset into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=42)

# Data augmentation for the training set
train_datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

# Data generator for the validation set (without augmentation)
val_datagen = ImageDataGenerator()

# Fit the data generator on your training set
train_datagen.fit(X_train)
val_datagen.fit(X_val)

# Train the model using the data generators
history = model.fit(
    train_datagen.flow(X_train, y_train, batch_size=32),
    validation_data=val_datagen.flow(X_val, y_val),
    epochs=20,
    callbacks=[early_stopping]
)

# Evaluate the model on the test set
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Save the trained model
model.save('drowsiness_detection_finetuned.h5')
```

```python
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()

# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
#checking only drowsy and non-drowsy
import cv2
import numpy as np

# Load the face and eye cascades
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')

# Open webcam
cap = cv2.VideoCapture(1)

# Check if the webcam is opened correctly
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
    if not cap.isOpened():
        raise IOError("Cannot open webcam")

status = "No status"

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Check if frame is captured successfully
    if not ret:
        print("Failed to capture image")
        break

    # Convert frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Detect eyes
    eyes = eye_cascade.detectMultiScale(gray, 1.1, 4)
    
    if len(eyes) > 0:  # If eyes are detected
        for (x, y, w, h) in eyes:
            roi_gray = gray[y:y+h, x:x+w]
            roi_color = frame[y:y+h, x:x+w]

            # Draw a rectangle around the eyes
            cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)

            # Extract the region of interest (eyes)
            eyes_roi = roi_color

            # Resize the eye ROI for the model
            final_image = cv2.resize(eyes_roi, (224, 224))
            final_image = np.expand_dims(final_image, axis=0)  # Add batch dimension
            final_image = final_image / 255.0  # Normalize the image

            # Make a prediction using the loaded model
            predictions = new_model.predict(final_image)

            # Update status based on prediction
            if predictions > 0.5:
                status = "Non-Drowsy"
            else:
                status = "Drowsy"
            print(f"Prediction: {predictions} | Status: {status}")
    else:
        status = "No Eyes Detected"

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Define the font for the text
    font = cv2.FONT_HERSHEY_SIMPLEX

    # Use putText() method to insert text on the video
    cv2.putText(frame, status, (58, 58), font, 1, (0, 0, 255), 2, cv2.LINE_AA)

    # Display the video frame
    cv2.imshow('Drowsiness Detection', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

# Release resources and close windows
cap.release()
cv2.destroyAllWindows()
```

```python
#car simulation 
import pygame

def run_car_simulation():
    pygame.init()

    # Set up the display
    screen = pygame.display.set_mode((800, 600))
    pygame.display.set_caption("Car Simulation")

    # Load the car and road images with the updated file paths
    try:
        car = pygame.image.load(r'C:\Users\Mayuri\Desktop\Driver Drowsiness (2)\Driver Drowsiness\car.png')
        road = pygame.image.load(r'C:\Users\Mayuri\Desktop\Driver Drowsiness (2)\Driver Drowsiness\road.jpg')  # Load the road image
    except pygame.error as e:
        print(f"Error loading image: {e}")
        return

    # Scale the road image to fit the entire window
    road = pygame.transform.scale(road, (800, 600))

    # Initial positions
    car_x, car_y = 330, 200  # Position the car at the bottom
    road_y = 0  # Start the road at the top
    speed = 10
    running = True

    # Set up clock for frame rate control
    clock = pygame.time.Clock()

    while running:
        # Handle events
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                running = False

        # Move the road vertically
        road_y += speed

        # If the road goes off the screen, reset its position
        if road_y > 0:
            road_y = -600  # Start from the top again

        # Clear the screen
        screen.fill((255, 255, 255))

        # Draw the road at the current position
        screen.blit(road, (0, road_y))
        screen.blit(road, (0, road_y + 600))  # Draw a second road image to cover the whole window

        # Draw the car on top of the road
        screen.blit(car, (car_x, car_y))

        # Update the display
        pygame.display.update()

        # Control the frame rate
        clock.tick(60)

    pygame.quit()

# Run the simulation with the car stationary and road moving
run_car_simulation()
```

```python
# main code slow lane change 
import winsound
import numpy as np
import cv2
import pygame
from tensorflow.keras.models import load_model

# Load the face and eye detection cascades
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye_tree_eyeglasses.xml')

# Frequency and duration for beep sound
frequency = 2500  # Set Frequency To 2500 Hertz
duration = 1000  # Set Duration To 1000 ms (1 second)

# Load the pre-trained drowsiness detection model
model_path = r'C:\Users\Mayuri\Desktop\drowsiness_detection_finetuned.h5'
new_model = load_model(model_path)

cap = cv2.VideoCapture(1)

# Check if the webcam is opened correctly
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

counter = 0
eyes_roi = None
threshold = 0.5  # This is the prediction threshold, adjust it as necessary

# Initialize the car simulation state variables
car_x, car_y = 330, 200  # Initial car position
speed = 10  # Initial car speed
drowsy_alerts = 0  # Count for drowsy alerts

# Set up pygame for car simulation
pygame.init()
screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Car Simulation")

# Load car and road images with full paths
try:
    car = pygame.image.load(r'C:\Users\Mayuri\Desktop\Driver Drowsiness (2)\Driver Drowsiness\car.png')
    road = pygame.image.load(r'C:\Users\Mayuri\Desktop\Driver Drowsiness (2)\Driver Drowsiness\road.jpg')
except pygame.error as e:
    print(f"Error loading image: {e}")
    pygame.quit()

# Scale the road image to fit the entire window
road = pygame.transform.scale(road, (800, 600))
road_y = 0  # Start the road at the top

# Set up clock for frame rate control
clock = pygame.time.Clock()

# Run the drowsiness detection and car simulation in the same loop
running = True
while running:
    ret, frame = cap.read()

    if not ret:
        print("Failed to grab frame")
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]

        # Detect eyes within the face region
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 4)

        for (ex, ey, ew, eh) in eyes:
            eyes_roi = roi_color[ey:ey + eh, ex:ex + ew]
            cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)

    # Check if eyes ROI is captured before further processing
    if eyes_roi is not None:
        try:
            # Resize and preprocess the eyes ROI
            final_image = cv2.resize(eyes_roi, (224, 224))
            final_image = np.expand_dims(final_image, axis=0)  # need fourth dimension
            final_image = final_image / 255.0

            # Predict using the loaded model
            predictions = new_model.predict(final_image)
            print(f"Model prediction: {predictions}")  # Debug output

            font = cv2.FONT_HERSHEY_SIMPLEX

            # Use a threshold to classify open or closed eyes
            if predictions[0][0] > threshold:
                status = "Non-Drowsy"
                counter = 0  # Reset the counter
                drowsy_alerts = 0  # Reset drowsy alerts when eyes are open
                speed = 10  # Normal speed
                cv2.putText(frame, status, (150, 150), font, 3, (0, 255, 0), 2, cv2.LINE_4)
            else:
                counter += 1
                status = "Drowsy"
                cv2.putText(frame, status, (150, 150), font, 3, (0, 0, 255), 2, cv2.LINE_4)

            # If eyes are closed for too long (drowsiness detected)
            if counter > 5:
                drowsy_alerts += 1
                speed = max(5, speed - 1)  # Decrease speed

                x1, y1, w1, h1 = 0, 0, 175, 75  # Coordinates for the "Sleep Alert" box
                cv2.rectangle(frame, (x1, y1), (x1 + w1, y1 + h1), (0, 0, 0), -1)

                # Add alert text
                cv2.putText(frame, 'Sleep Alert !!', (x1 + int(w1 / 10), y1 + int(h1 / 2)),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                winsound.Beep(frequency, duration)
                counter = 0  # Reset the counter after alert

                # If too many alerts, move to the left lane and stop
                if drowsy_alerts > 5:
                    if car_x > 100:  # Move car to the left lane gradually
                        car_x -= 5
                    else:
                        speed = 0  # Stop the car

        except Exception as e:
            print(f"Error during prediction or processing: {e}")

    # --- Car simulation ---
    # Handle pygame events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Move the road vertically
    road_y += speed

    # If the road goes off the screen, reset its position
    if road_y > 0:
        road_y = -600  # Start from the top again

    # Clear the screen
    screen.fill((255, 255, 255))

    # Draw the road at the current position
    screen.blit(road, (0, road_y))
    screen.blit(road, (0, road_y + 600))  # Draw a second road image to cover the whole window

    # Draw the car on top of the road
    screen.blit(car, (car_x, car_y))

    # Update the display
    pygame.display.update()

    # Control the frame rate
    clock.tick(60)

    cv2.imshow('Drowsiness Detection Tutorial', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        running = False

cap.release()
cv2.destroyAllWindows()
pygame.quit()
```

```python
#main code fast lane change 
import winsound
import numpy as np
import cv2
import pygame
from tensorflow.keras.models import load_model

# Load the face and eye detection cascades
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye_tree_eyeglasses.xml')

# Frequency and duration for beep sound
frequency = 2500  # Set Frequency To 2500 Hertz
duration = 1000  # Set Duration To 1000 ms (1 second)

# Load the pre-trained drowsiness detection model
model_path = r'C:\Users\Mayuri\Desktop\drowsiness_detection_finetuned.h5'  # Update this path
new_model = load_model(model_path)

cap = cv2.VideoCapture(1)

# Check if the webcam is opened correctly
if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open webcam")

counter = 0
eyes_roi = None
threshold = 0.5  # This is the prediction threshold, adjust it as necessary

# Initialize the car simulation state variables
car_x, car_y = 330, 200  # Initial car position
speed = 15  # Start with a slightly fast initial car speed
drowsy_alerts = 0  # Count for drowsy alerts

# Set up pygame for car simulation
pygame.init()
screen = pygame.display.set_mode((800, 600))
pygame.display.set_caption("Car Simulation")

# Load car and road images with full paths
try:
    car = pygame.image.load(r'C:\Users\Mayuri\Desktop\Driver Drowsiness (2)\Driver Drowsiness\car.png')
    road = pygame.image.load(r'C:\Users\Mayuri\Desktop\Driver Drowsiness (2)\Driver Drowsiness\road.jpg')
except pygame.error as e:
    print(f"Error loading image: {e}")
    pygame.quit()

# Scale the road image to fit the entire window
road = pygame.transform.scale(road, (800, 600))
road_y = 0  # Start the road at the top

# Set up clock for frame rate control
clock = pygame.time.Clock()

# Run the drowsiness detection and car simulation in the same loop
running = True
while running:
    ret, frame = cap.read()

    if not ret:
        print("Failed to grab frame")
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Draw a rectangle around the faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        roi_gray = gray[y:y + h, x:x + w]
        roi_color = frame[y:y + h, x:x + w]

        # Detect eyes within the face region
        eyes = eye_cascade.detectMultiScale(roi_gray, 1.1, 4)

        for (ex, ey, ew, eh) in eyes:
            eyes_roi = roi_color[ey:ey + eh, ex:ex + ew]
            cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)

    # Check if eyes ROI is captured before further processing
    if eyes_roi is not None:
        try:
            # Resize and preprocess the eyes ROI
            final_image = cv2.resize(eyes_roi, (224, 224))
            final_image = np.expand_dims(final_image, axis=0)  # need fourth dimension
            final_image = final_image / 255.0

            # Predict using the loaded model
            predictions = new_model.predict(final_image)
            print(f"Model prediction: {predictions}")  # Debug output

            font = cv2.FONT_HERSHEY_SIMPLEX
            text_position = (frame.shape[1] - 300, 50)  # Adjust coordinates to top-right corner

            # Use a threshold to classify open or closed eyes
            if predictions[0][0] > threshold:
                status = "Non-Drowsy"
                counter = 0  # Reset the counter
                drowsy_alerts = 0  # Reset drowsy alerts when eyes are open
                speed = 15  # Reset to fast speed if eyes are open
                cv2.putText(frame, status, text_position, font, 1, (0, 255, 0), 2, cv2.LINE_4)
            else:
                counter += 1
                status = "Drowsy"
                cv2.putText(frame, status, text_position, font, 1, (0, 0, 255), 2, cv2.LINE_4)

            # If eyes are closed for too long (drowsiness detected)
            if counter > 5:
                drowsy_alerts += 1
                speed = max(5, speed - 1)  # Decrease speed

                x1, y1, w1, h1 = 0, 0, 175, 75  # Coordinates for the "Sleep Alert" box
                cv2.rectangle(frame, (x1, y1), (x1 + w1, y1 + h1), (0, 0, 0), -1)

                # Add alert text
                cv2.putText(frame, 'Sleep Alert !!', (x1 + int(w1 / 10), y1 + int(h1 / 2)),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                winsound.Beep(frequency, duration)
                counter = 0  # Reset the counter after alert

                # If too many alerts, move to the left lane and stop
                if drowsy_alerts >= 3:  # After 3 alerts, move car to the left lane and stop
                    if car_x > 100:  # Move car to the left lane gradually
                        car_x -= 20  # Increase the speed of moving left
                    else:
                        speed = 0  # Stop the car

        except Exception as e:
            print(f"Error during prediction or processing: {e}")

    # --- Car simulation ---
    # Handle pygame events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Move the road vertically
    road_y += speed

    # If the road goes off the screen, reset its position
    if road_y > 0:
        road_y = -600  # Start from the top again

    # Clear the screen
    screen.fill((255, 255, 255))

    # Draw the road at the current position
    screen.blit(road, (0, road_y))
    screen.blit(road, (0, road_y + 600))  # Draw a second road image to cover the whole window

    # Draw the car on top of the road
    screen.blit(car, (car_x, car_y))

    # Update the display
    pygame.display.update()

    # Control the frame rate
    clock.tick(60)

    cv2.imshow('Drowsiness Detection Tutorial', frame)

    if cv2.waitKey(2) & 0xFF == ord('q'):
        running = False

cap.release()
cv2.destroyAllWindows()
pygame.quit()
```

```python
# counting total images
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Initialize ImageDataGenerator
datagen = ImageDataGenerator()

# Load data using flow_from_directory
train_data = datagen.flow_from_directory(r'C:\Users\Mayuri\Desktop\Driver Drowsiness (2)\Driver Drowsiness\Train_dataset')
test_data = datagen.flow_from_directory(r'C:\Users\Mayuri\Desktop\Driver Drowsiness (2)\Driver Drowsiness\Test_dataset')

# Get the number of images
train_images_count = train_data.samples
test_images_count = test_data.samples
total_images_count = train_images_count + test_images_count

# Print the results
print(f'Total Train Images: {train_images_count}')
print(f'Total Test Images: {test_images_count}')
print(f'Total Dataset Images: {total_images_count}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define image dimensions (update these values as per your model's requirements)
image_height = 224  # Example height
image_width = 224   # Example width

# Load the trained model from the specified path
model_path = r'C:\Users\Mayuri\Desktop\drowsiness_detection_finetuned.h5'
new_model = load_model(model_path)

# Define paths to your test dataset
test_data_path = r'C:\Users\Mayuri\Desktop\Driver Drowsiness (2)\Driver Drowsiness\Test_dataset'

# Create an ImageDataGenerator for preprocessing
test_datagen = ImageDataGenerator(rescale=1.0/255)  # Normalizing the images

# Load test data
test_generator = test_datagen.flow_from_directory(
    test_data_path,
    target_size=(image_height, image_width),  # Use the defined dimensions
    batch_size=1,
    class_mode='binary',  # Assuming binary classification
    shuffle=False  # Keep the order for evaluation
)

# Getting the test images and labels
X_test, y_test = [], []
for i in range(len(test_generator)):
    img, label = test_generator[i]  # Get the next batch of images and labels
    X_test.append(img[0])  # Add the single image
    y_test.append(label[0])  # Add the label

# Convert to numpy arrays
X_test = np.array(X_test)
y_test = np.array(y_test)

# Make predictions on the test set
y_pred = new_model.predict(X_test)

# Convert predicted probabilities to binary class labels (0 or 1)
y_pred_classes = np.where(y_pred > 0.5, 1, 0)  # Assuming binary classification

# Generate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred_classes)

# Define class names (modify based on your dataset)
Classes = ['Non_Drowsy', 'Drowsy']

# Plot the confusion matrix using Seaborn's heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False, 
            xticklabels=Classes, yticklabels=Classes, linewidths=1, linecolor='black')

# Set the labels and title for the plot
plt.xlabel('Predicted Labels', fontsize=12)
plt.ylabel('True Labels', fontsize=12)
plt.title('Confusion Matrix', fontsize=15)

# Display the plot
plt.show()

# Print classification report with zero_division set to avoid warnings
print(classification_report(y_test, y_pred_classes, target_names=Classes, zero_division=1))
```
